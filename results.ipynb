```python
import matplotlib.pyplot as plt
import csv
import numpy as np

# Load the CSV data
with open('parent-data-2023-05-05.csv') as f:
    reader = csv.reader(f)
    header_row = next(reader)
    parents_edu = []
    math_score = []
    for row in reader:
        parents_edu.append(row[3])
        math_score.append(int(row[6]))   


# Create four bins for the math scores
bins = np.linspace(min(math_score), max(math_score), 5)

# Group the math scores into four bins
math_score_groups = np.digitize(math_score, bins)

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(10, 6))

# Create a scatter plot with colored groups
ax.scatter(parents_edu, math_score, c=math_score_groups, cmap='cool')

# Set the x-label and y-label
ax.set_xlabel('Parents Education')
ax.set_ylabel('Math Score')

# Set the title
ax.set_title('Correlation between Parents Education and Math Score')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("alc-data-2023-05-07.csv", delimiter=",")

# Calculate the average grade
data["avg_grade"] = (data["G1"] + data["G2"] + data["G3"]) / 3
data["avg_alc"] = round((data["Walc"] + data["Dalc"]) / 2 )

# Plot the data
plt.bar(data["avg_alc"], data["avg_grade"])
plt.title("Overall Alcohol Consumption vs. Average Grade")
plt.xlabel("Overall Alcohol Consumption")
plt.ylabel("Average Grade")
plt.show()
```
